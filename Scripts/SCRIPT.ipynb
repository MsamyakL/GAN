```python
import os 
import shutil
```

```python
src = 'C:\\Users\\samya\\Downloads\\DEFICIENCY'
dest1 = 'C:\\Users\\samya\\Downloads\\Chan1\\jpg'
dest2 = 'C:\\Users\\samya\\Downloads\\Chan1\\png'
dest3 = 'C:\\Users\\samya\\Downloads\\Chan1\\jpeg'
```

```python
for dirs,subdirs,files in os.walk(src):
    
    for file in files:
        if file.endswith('.jpg'):
            print(file)

```

```python
for dirs,subdirs,files in os.walk(src):
    
    for file in files:
        if file.endswith('.png'):
            
            filename = os.path.join(src,dirs,file)
            if os.path.exists(filename):
                print(filename)
                
                shutil.copy(filename,dest2)
```
